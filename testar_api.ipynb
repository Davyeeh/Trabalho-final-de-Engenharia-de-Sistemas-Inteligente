# Lendo dataset

```python
import requests
import pandas as pd
import os

# Configura√ß√£o
BASE_URL = "http://127.0.0.1:8000"
CAMINHO_CSV = "dataset/dataset_original.csv"  # Caminho relativo do ficheiro

# 1. Verificar e Visualizar os dados antes de enviar
if os.path.exists(CAMINHO_CSV):
    print(f"üìÇ Ficheiro encontrado: {CAMINHO_CSV}")
    try:
        # Apenas para visualiza√ß√£o no Notebook
        df_preview = pd.read_csv(CAMINHO_CSV)
        print(f"Total de registos: {len(df_preview)}")
        display(df_preview.head(3))
    except Exception as e:
        print(f"Aviso: N√£o foi poss√≠vel ler o preview com pandas: {e}")
else:
    print(f"‚ùå ERRO: Ficheiro n√£o encontrado em {CAMINHO_CSV}")
    # Pare a execu√ß√£o se n√£o achar o ficheiro
    raise FileNotFoundError("Ficheiro de dataset n√£o encontrado.")
```

# Treinando modelo 

```python
endpoint_train = f"{BASE_URL}/train"

print("\n‚è≥ A enviar ficheiro para treino...")

# Abrimos o ficheiro em modo de leitura bin√°ria ('rb')
with open(CAMINHO_CSV, 'rb') as f:
    files = {'file': (os.path.basename(CAMINHO_CSV), f, 'text/csv')}
    
    try:
        response = requests.post(endpoint_train, files=files)
        
        if response.status_code == 200:
            resultado = response.json()
            print("\n‚úÖ Treino conclu√≠do com sucesso!")
            print("-" * 30)
            print(f"üìä R¬≤ Score: {resultado['metricas_R2']}")
            print(f"üìù Colunas usadas: {resultado['colunas_usadas']}")
            print("-" * 30)
        else:
            print(f"\n‚ùå Erro na API ({response.status_code}):")
            print(response.text)

    except Exception as e:
        print(f"\n‚ùå Falha na conex√£o com a API: {e}")
```

# Fazendo predi√ß√£o

```python
# C√©lula de Teste de Predi√ß√£o (/predict)
endpoint_predict = f"{BASE_URL}/predict"

# Exemplo de payload (ajuste conforme um caso real do seu dataset)
payload = {
    "area": 50,
    "bedrooms": 2,
    "bathrooms": 2,
    "parking_spaces": 1,
    "neighborhood": "Pinheiros", 
    "latitude": -23.56413595,
    "longitude": -46.69517321540903,
    "created_date": 2020
}

print(f"üîÆ A estimar pre√ßo para im√≥vel em {payload['neighborhood']}...")

try:
    response = requests.post(endpoint_predict, json=payload)
    
    if response.status_code == 200:
        predicao = response.json()
        valor = predicao['preco_estimado']
        print(f"\nüí∞ Pre√ßo Estimado: R$ {valor:,.2f}")
    else:
        print(f"Erro: {response.text}")
except Exception as e:
    print(f"Erro: {e}")
```
